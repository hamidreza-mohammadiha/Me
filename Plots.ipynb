```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function R(n)
def R(n):
    return (n * np.sin(2 * np.pi / n)) / (2 * np.pi)

def P(n):
    return (n) / ((2 * np.pi - 4) + (n))

def Q(n):
    return (n) / ((2 * np.sqrt(2)*np.pi - 8) + (n))

def P(n, k=0.42, n0=2):
    return 1 - np.exp(-k * (n - n0))

# Generate values for n
n_values = np.arange(0, 21)  # start from 3 for the polygon (triangle)
R_values = R(n_values)
P_values = P(n_values)
Q_values = Q(n_values)
P_values = P(n_values)  # Exponential approximation

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(n_values, R_values, label='R(n) = n * sin(2/n) / (2)', color='blue')
# plt.plot(n_values, P_values, label='P(n) = n / ((2-4)+n)', color='red')
# plt.plot(n_values, Q_values, label='P(n) = n / ((2*2^-1*-8)+n))', color='green')
# plt.axhline(y=2/np.pi, color='orange', linestyle='--', label='R(4) = 2/', linewidth=1.5)
# plt.axhline(y=(2 * np.sqrt(2)/np.pi), color='orange', linestyle='--', label='R(4) = 2.2^(0.5))/', linewidth=1.5)
plt.plot(n_values, P_values, label='P(n) = 1 - e^(-0.42(n - 2))', color='black', linestyle='--')
plt.title('Ratio of the Area of Polygon to Area of Circle')
plt.xlabel('Number of Sides (n)')
plt.ylabel('R(n)')
plt.grid()
plt.legend()
plt.xlim(0, 21)
plt.ylim(0, 1)
plt.xticks(np.arange(0, 21, 1))
plt.yticks(np.linspace(0, 1, 22))
plt.show()
```
